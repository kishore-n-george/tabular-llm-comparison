# Section 2: Model Training and Evaluation

## Bike Sharing Dataset - XGBoost Regression Model

This notebook trains and evaluates XGBoost regression model on the Bike Sharing Dataset.

**Model Evaluated:**
- **XGBoost**: Gradient boosting with tree-based learners for regression

**Evaluation Components:**
- Comprehensive regression performance metrics
- Actual vs Predicted plots
- Residual analysis
- Feature importance analysis
- Cross-validation evaluation
- Computational efficiency analysis

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings
import os
warnings.filterwarnings('ignore')

# Import our XGBoost training functions
from xgboost_training_functions import (
    setup_logging,
    load_preprocessed_data,
    initialize_evaluator,
    train_xgboost_model,
    analyze_feature_importance,
    analyze_model_performance,
    perform_cross_validation,
    analyze_predictions_by_range,
    perform_error_analysis,
    generate_business_insights,
    save_results,
    print_final_summary
)

# Set random seeds for reproducibility
np.random.seed(42)

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("üöÄ Section 2: XGBoost Model Training and Evaluation")
print("Dataset: Bike Sharing Dataset")
print("Task: Regression - Predicting bike rental count")
```

## 2.1 Setup Logging and Load Data

```python
# Setup logging - all print statements will be logged to Section2_Model_Training.log
logger = setup_logging('Section2_Model_Training.log')

# Load preprocessed data from Section 1
data = load_preprocessed_data(logger)

print("‚úÖ Data loaded and logging setup completed!")
```

## 2.2 Initialize Evaluator

```python
# Initialize comprehensive evaluator
evaluator, models = initialize_evaluator(data, logger)

print("‚úÖ Evaluator initialized successfully!")
```

## 2.3 Train XGBoost Model

```python
# Train XGBoost regression model
xgb_model, xgb_results = train_xgboost_model(data, evaluator, logger)
models['XGBoost'] = xgb_model

print("‚úÖ XGBoost model trained successfully!")
```

## 2.4 Feature Importance Analysis

```python
# Analyze and plot feature importance
feature_importance_df = analyze_feature_importance(xgb_model, data['feature_names'], logger)

print("‚úÖ Feature importance analysis completed!")
```

## 2.5 Model Performance Analysis

```python
# Analyze model performance in detail
y_pred, performance_metrics = analyze_model_performance(xgb_model, data, logger)

print("‚úÖ Model performance analysis completed!")
```

## 2.6 Cross-Validation Analysis

```python
# Perform k-fold cross-validation
cv_scores = perform_cross_validation(xgb_model, data, logger)

print("‚úÖ Cross-validation analysis completed!")
```

## 2.7 Prediction Analysis by Range

```python
# Analyze prediction accuracy across different target value ranges
analyze_predictions_by_range(data['y_test'], y_pred, logger)

print("‚úÖ Prediction range analysis completed!")
```

## 2.8 Error Analysis

```python
# Perform detailed error analysis
residuals, abs_residuals = perform_error_analysis(data['y_test'], y_pred, logger)

print("‚úÖ Error analysis completed!")
```

## 2.9 Business Insights

```python
# Generate business insights for bike sharing
generate_business_insights(feature_importance_df, data['feature_names'], logger)

print("‚úÖ Business insights generated!")
```

## 2.10 Save Results

```python
# Save all results and models
save_results(xgb_model, xgb_results, feature_importance_df, data['y_test'], y_pred, residuals, abs_residuals, logger)

print("‚úÖ All results saved successfully!")
```

## 2.11 Final Summary

```python
# Print comprehensive final summary
print_final_summary(performance_metrics, cv_scores, feature_importance_df, data, logger)

print("\nüéâ XGBoost training and evaluation completed successfully!")
```

```python
# Display final results summary
print("\n" + "="*80)
print("FINAL RESULTS SUMMARY")
print("="*80)

print(f"\nüìä XGBoost Performance:")
print(f"   R¬≤ Score: {performance_metrics['r2']:.4f}")
print(f"   RMSE: {performance_metrics['rmse']:.4f}")
print(f"   MAE: {performance_metrics['mae']:.4f}")
print(f"   MAPE: {performance_metrics['mape']:.4f}%")

print(f"\nüîÑ Cross-Validation:")
print(f"   Mean R¬≤ Score: {cv_scores.mean():.4f} ¬± {cv_scores.std():.4f}")

print(f"\nüéØ Top 3 Most Important Features:")
for i, (_, row) in enumerate(feature_importance_df.head(3).iterrows(), 1):
    print(f"   {i}. {row['feature']}: {row['importance']:.3f}")

print(f"\nüìÅ Generated Files:")
print(f"   - Section2_Model_Training/xgboost_evaluation_results.csv")
print(f"   - Section2_Model_Training/xgboost_feature_importance.csv")
print(f"   - Section2_Model_Training/xgboost_predictions.csv")
print(f"   - Section2_Model_Training/xgboost_model.pkl")
print(f"   - Section2_Model_Training/XGBoost_feature_importance.png")
print(f"   - Section2_Model_Training/XGBoost_regression_results.png")
print(f"   - Section2_Model_Training/XGBoost_residual_analysis.png")
print(f"   - Section2_Model_Training/Section2_Model_Training.log")

print(f"\nüöÄ Model ready for deployment and comparison with other models!")
```

## 2.13 Quick Model Test

```python
# Quick test of the trained model
print("\nüß™ Quick Model Test:")
print("-" * 30)

# Test on a few samples
test_indices = [0, 100, 500, 1000, 2000]
print(f"{'Index':<6} {'Actual':<8} {'Predicted':<10} {'Error':<8}")
print("-" * 35)

for idx in test_indices:
    if idx < len(data['y_test']):
        actual = data['y_test'][idx]
        predicted = y_pred[idx]
        error = abs(actual - predicted)
        print(f"{idx:<6} {actual:<8.0f} {predicted:<10.1f} {error:<8.1f}")

print(f"\n‚úÖ Model is working correctly!")
```
